```python
# Import the Lib
import pandas as pd
import numpy as np
```

```python
#Reading the dataset
url = 'https://raw.githubusercontent.com/SnehaDebnath678/number_of_order_prediction/refs/heads/main/supplement.csv'
dataset = pd.read_csv(url)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
# pip install plotly
import plotly.express as px
```

```python
pie = dataset["Store_Type"].value_counts()
store = pie.index
orders = pie.values

fig = px.pie(dataset, values=orders, names=store)
fig.show()
```

```python
pie2 = dataset["Location_Type"].value_counts()
location = pie2.index
orders = pie2.values

fig = px.pie(dataset, values=orders, names=location)
fig.show()
```

```python
pie3 = dataset["Discount"].value_counts()
discount = pie3.index
orders = pie3.values

fig = px.pie(dataset, values=orders, names=discount)
fig.show()
```

```python
pie4 = dataset["Holiday"].value_counts()
holiday = pie4.index
orders = pie4.values

fig = px.pie(dataset, values=orders, names=holiday)
fig.show()
```

```python
dataset["Discount"] = dataset["Discount"].map({"No": 0,"Yes": 1})
```

```python
dataset
```

```python
dataset["Store_Type"] = dataset["Store_Type"].map({"S1": 1, "S2": 2,"S3": 3,"S4": 4})
```

```python
dataset
```

```python
dataset["Location_Type"] = dataset["Location_Type"].map({"L1": 1, "L2": 2,"L3": 3,"L4": 4})
```

```python
dataset
```

```python
x = np.array(dataset[["Store_Type", "Location_Type", "Holiday", "Discount"]])
y = np.array(dataset["#Order"])
```

```python
x
```

```python
y
```

```python
# Building the ML model
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
len(x_train)
```

```python
# pip install dask[dataframe]
```

```python
# pip install lightgbm
import lightgbm as ltb
model = ltb.LGBMRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```
